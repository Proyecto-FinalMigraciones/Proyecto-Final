# ETL

```python
import pandas as pd
```

# DESTINO MUJERES

```python
destino_mujeres = pd.read_csv('Proyecto-Final/Datasets/destino_mujeres.csv', skiprows = 10)
destino_mujeres
```

```python
destino_mujeres.drop(['Unnamed: 0'], axis=1, inplace = True)
destino_mujeres
```

```python
destino_mujeres.info()
```

```python
destino_mujeres['Type of data'] = destino_mujeres['Type of data'].fillna('0')
```

```python
destino_mujeres.info()
```

```python
columnas = ['1990', '1995', '2000', '2005', '2010', '2015', '2020']
```

```python
for col in columnas:
    destino_mujeres[col] = pd.to_numeric(destino_mujeres[col], errors='coerce').fillna(0)
```

```python
destino_mujeres
```

```python
destino_mujeres['Notes'] = destino_mujeres['Notes'].fillna('0')
```

```python
destino_mujeres.info()
```

```python
year_columns = destino_mujeres.columns[4:]

destino_mujeres = pd.melt(destino_mujeres, id_vars=destino_mujeres.columns[:4], value_vars=year_columns, var_name='Year', value_name='Numero de Migrantes')
```

```python
destino_mujeres
```

```python
valores_unicos = destino_mujeres['Region, development group, country or area'].unique()
print(valores_unicos)
```

```python
valores_a_eliminar = ['WORLD', 'Sub-Saharan Africa', 'Northern Africa and Western Asia',
                     'Central and Southern Asia', 'Eastern and South-Eastern Asia',
                     'Latin America and the Caribbean',
                     'Oceania (excluding Australia and New Zealand)',
                     'Australia and New Zealand', 'Europe and Northern America',
                     'Developed regions', 'Less developed regions',
                     'Less developed regions, excluding least developed countries',
                     'Less developed regions, excluding China',
                     'Least developed countries',
                     'Land-locked Developing Countries (LLDC)',
                     'Small island developing States (SIDS)',
                     'High-income countries', 'Middle-income countries',
                     'Upper-middle-income countries', 'Lower-middle-income countries',
                     'Low-income countries', 'AFRICA', 'Eastern Africa']
```

```python
destino_mujeres_pais = destino_mujeres[~destino_mujeres['Region, development group, country or area'].str.strip().isin(valores_a_eliminar)].copy()
destino_mujeres_pais
```

```python
destino_mujeres_region =  destino_mujeres.head(23).copy()
```

```python
destino_mujeres_region
```

```python
destino_mujeres
```

```python
destino_mujeres_pais.to_csv('destino_mujeres_pais.csv', index = False)
```

```python
destino_mujeres_region.to_csv('destino_mujeres_region.csv', index = False)
```

# DESTINO PORCENTAJE

```python
destino_porcentaje = pd.read_csv('Proyecto-Final/Datasets/destino_porcentaje.csv', skiprows = 10)
destino_porcentaje
```

```python
destino_porcentaje.drop(['Unnamed: 0'], axis=1, inplace = True)
destino_porcentaje
```

```python
destino_porcentaje.info()
```

```python
destino_porcentaje[['1990', '1990.1', '1990.2']]
```

```python
destino_porcentaje['Notes'] = destino_porcentaje['Notes'].fillna('0')
```

```python
destino_porcentaje['Type of data'] = destino_porcentaje['Type of data'].fillna('0')
destino_porcentaje.info()
```

```python
columnas = ['1990', '1995', '2000', '2005', '2010', '2015', '2020', '1990.1', '1995.1', '2000.1', '2005.1', '2010.1', '2015.1', '2020.1', '1990.2', '1995.2', '2000.2', '2005.2', '2010.2', '2015.2', '2020.2']
```

```python
for col in columnas:
    destino_porcentaje[col] = pd.to_numeric(destino_porcentaje[col], errors='coerce').fillna(0)
```

```python
destino_porcentaje.info()
```

```python
(destino_porcentaje['1990'] == 0).sum()
```

```python
columnas_excluidas = ['Region, development group, country or area', 'Notes', 'Location code', 'Type of data']
```

```python
for col in destino_porcentaje.columns:
    if col not in columnas_excluidas:
        if col.endswith('.1'):
            new_col = col.replace('.1', 'H')
        elif col.endswith('.2'):
            new_col = col.replace('.2', 'M')
        else:
            new_col = col + 'C'
        destino_porcentaje.rename(columns={col: new_col}, inplace=True)
```

```python
destino_porcentaje
```

```python
valores_unicosp = destino_porcentaje['Region, development group, country or area'].unique()
print(valores_unicosp)
```

```python
destino_porcentaje_pais = destino_porcentaje[~destino_porcentaje['Region, development group, country or area'].str.strip().isin(valores_a_eliminar)].copy()
destino_porcentaje_pais
```

```python
destino_porcentaje_region =  destino_porcentaje.head(23).copy()
destino_porcentaje_region
```

```python
year_columns = destino_porcentaje_pais.columns[4:]

destino_porcentaje_pais = pd.melt(destino_porcentaje_pais, id_vars=destino_porcentaje_pais.columns[:4], value_vars=year_columns, var_name='Year', value_name='Numero de Migrantes')
```
